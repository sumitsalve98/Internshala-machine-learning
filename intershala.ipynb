## The pandas.apply() Function

```python
data.info()
```

```python
"""
problem statement:
A house can be called a luxury home is any two of the three conditions are met.
1. 'Waterfront View' is present
2. 'Condition of the House'is either good or excellent
3. 'Overall Grade' is greater than or equal to 8
"""
```

```python
# by looping
data['luxury_home'] = 0

for i in range(len(data)):
    count = 0
    if data['Waterfront View'][i] == 'Yes':
        count += 1
    if data['Condition of the House'][i] in ['Good', 'Excellent']:
        count += 1
    if data['Overall Grade'][i] >=8:
        count += 1
    if count >= 2:
        data['luxury_home'][i] = 'yes'
    else:
        data['luxury_home'][i] = 'No'
```

```python
data.head()
```

```python
def luxury_home(row):
    count = 0
    if row[0] == 'Yes':
        count += 1
    if row[1] in ['Good', 'Excellent']:
        count += 1
    if row[2] >=8:
        count += 1
    if count >= 2:
        return 'yes'
    else:
        return 'No'
data['luxury_home'] = data[['Waterfront View','Condition of the House','Overall Grade']].apply(luxury_home, axis = 1)
```
